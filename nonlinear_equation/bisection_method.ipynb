### Definisikan Fungsi Bisection Method

```python
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    i = 0
    error = float('inf')
    print("Iterasi\t a\t\t b\t\t c\t\t f(c)\t\t Error")
    while error > tol and i < max_iter:
        c = (a + b) / 2
        fc = f(c)
        error = abs(fc)
        print(f"{i+1}\t {a:.6f}\t {b:.6f}\t {c:.6f}\t {fc:.6f}\t {error:.6f}")
        if fc == 0:
            break
        elif f(a) * fc < 0:
            b = c
        else:
            a = c
        i += 1
    return c
```

### Contoh 1: 
* Diberikan sistem persamaan: 
$$f(x)=x^3-3x++1$$
* Dengan interval: 
$$a=0, b=1$$

```python
def f1(x):
    return x**3 - 3*x + 1

root = bisection_method(f1, 0, 1)
print(f"Akar yang ditemukan: {root:.6f}")
```

### Contoh 2: 
* Diberikan sistem persamaan: 
$$ x-cos(x)=0 $$
* Dengan interval: 
$$a=0.6, b=0.8$$

```python
def f2(x):
    return x- np.cos(x)
root = bisection_method(f2, 0, 1, tol=1e-13)
print(f"Akar yang ditemukan: {root:.13f}")
```
