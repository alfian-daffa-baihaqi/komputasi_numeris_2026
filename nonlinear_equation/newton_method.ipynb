# NEWTON METHOD

```python
import numpy as np
```

### ONE DIMENSIONAL

```python
def newton_method_1d(f, df, x0, tol=1e-6, max_iter=100):
    i = 0
    error = np.inf
    print("Iter\t x\t\t f(x)\t\t f'(x)\t\t x_new\t\t Error")
    x = x0
    while error > tol and i < max_iter:
        f_val = f(x)
        df_val = df(x)
        if abs(df_val) < 1e-12:
            print("Derivative is too close to zero.")
            return None
        x_new = x - f_val / df_val
        error = abs(x_new - x)
        print(f"{i+1}\t {x:.6f}\t {f_val:.6f}\t {df_val:.6f}\t {x_new:.6f}\t {error:.6f}")
        x = x_new
        i += 1
    return x
```

#### Contoh 1: 
* Diberikan sistem persamaan: 
$$f(x)=x^3 - 2x^2 + x - 3$$
* Dengan initial value: 
$$x_0=4$$

```python
def f(x):
    return x**3 - 2*x**2 + x - 3
def df(x):
    return 3*x**2 - 4*x + 1
root = newton_method_1d(f, df, 4)
print("Akar persamaan nonlinear:", root)
```

### Contoh 2: 
* Diberikan sistem persamaan: 
$$ x-cos(x)=0 $$
* Dengan interval: 
$$a=0.6, b=0.8$$

```python
def f1(x):
    return x - np.cos(x)

def df1(x):
    return 1 + np.sin(x)
root1 = newton_method_1d(f1, df1, 0.8, tol=1e-13)
print("Akar persamaan nonlinear:", root1)
```

## Implementation of Newton Method for System of Equation (2D)

```python
def newton_method_2d(F, J, x0, y0):
    X = np.array([x0, y0], dtype=float)
    tol = 1e-6
    max_iter = 100
    i = 0
    error = np.inf
    print("Iter\t x\t\t y\t\t Error")
    while error > tol and i < max_iter:
        x,y = X
        F_val = F(x,y) 
        J_val = J(x,y)
        J_inv = np.linalg.inv(J_val)
        X_new = X - J_inv @ F_val
        error = np.linalg.norm(X_new - X)
        X = X_new
        print(f"{i+1}\t {X[0]:.6f}\t {X[1]:.6f}\t {error:.6f}")
        i += 1
    return X
```

#### Contoh 1
* Penyelesaian Sistem Persamaan Nonlinier
* Diberikan sistem persamaan nonlinier:

$$
f_1(x,y) = y + x^2 - 0.5 - x = 0
$$

$$
f_2(x,y) = x^2 - 5xy - y = 0
$$
* Tebakan Awal
$$
x_0 = 1, y_0 = 0
$$

```python
# Definisikan fungsi dan turunannya
def F2(x,y):
    return np.array([y + x**2 -1 -x, 
                     x**2 - 2*y**2 - y])
def J2(x,y):
    return np.array([[2*x - 1, 1], 
                     [2*x, -4*y - 1]])

root = newton_method_2d(F2, J2, 0, 0)
print("Akar persamaan nonlinear:", root)
```
