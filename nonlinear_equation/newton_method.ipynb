
Implementation of Newton Method for System of Equation

```python
import numpy as np
```

```python
# Definisikan fungsi dan turunannya
def F(x,y):
    return np.array([y + x**2 -1 -x, 
                     x**2 - 2*y**2 - y])
def J(x,y):
    return np.array([[2*x - 1, 1], 
                     [2*x, -4*y - 1]])
```

```python
def newton_method(x0, y0):
    X = np.array([x0, y0], dtype=float)
    tol = 1e-6
    max_iter = 100
    i = 0
    error = np.inf
    while error > tol and i < max_iter:
        x,y = X
        F_val = F(x,y) 
        J_val = J(x,y)
        J_inv = np.linalg.inv(J_val)
        X_new = X - J_inv @ F_val
        error = np.linalg.norm(X_new - X)
        X = X_new
        print(f"Iterasi {i+1}: X = {X}, Error = {error}")
        i += 1
    return X
```

```python
root = newton_method(0, 0)
print("Akar persamaan nonlinear:", root)
```
