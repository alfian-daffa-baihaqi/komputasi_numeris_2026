```python
# Persamaan dan Turunannya
def f(x):
  return x**2-2

def df(x):
  return 2*x
```

```python
def newton_raphson(f, df, x0):
  tol = 1e-6
  max_iter = 10
  i = 0
  x = x0
  error = float("inf")
  while (i < max_iter) and (error > tol):
    x_new = x - f(x)/df(x)
    error = abs(x_new - x)
    x = x_new
    print(x)
    i += 1

  return x
```

```python
root = newton_raphson(f, df, 5)
print("Akar persamaan adalah:", root)
```
