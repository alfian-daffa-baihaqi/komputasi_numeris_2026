# Gauss Jordan

```python
import numpy as np

def gauss_jordan_elimination(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)

    print("Augmented matrix awal:")
    print(np.column_stack((A, b)))

    for i in range(n):
        # Normalisasi pivot
        pivot = A[i, i]
        A[i] /= pivot
        b[i] /= pivot
        print(f"\nR{i+1} = R{i+1} / {pivot}")

        # Eliminasi baris lain
        for j in range(n):
            if j != i:
                factor = A[j, i]
                A[j] -= factor * A[i]
                b[j] -= factor * b[i]
                print(f"R{j+1} = R{j+1} - ({factor})R{i+1}")

        print(np.column_stack((A, b)))

    print("\nHasil akhir:")
    print(np.column_stack((A, b)))

    return b
```

### Contoh 1
Diberikan sistem persamaan dalam bentuk matriks
$$A= \begin{bmatrix}
 2 & -2 & 2  \\
 4 & 2 & -1 \\
 2 & -2 & 4
\end{bmatrix},
b= \begin{bmatrix}
 0 \\
 7 \\
 2 \\
\end{bmatrix}
$$

```python
A1 = np.array([[2, -2, 2],
               [4, 2, -1],
               [2, -2, 4]
              ])
b1 = np.array([0, 7, 2]) 

x = gauss_jordan_elimination(A1, b1)

print("\nSolusi x:")
print(x)
```
