## Golden Section

```python
import numpy as np
```

```python
def golden_section(f, a, b, tol=1e-5, max_iter=100):
    p = (3 - np.sqrt(5)) / 2  # Golden ratio conjugate
    c = a + p * (b - a)
    d = a + (1 - p) * (b - a)
    fc = f(c)
    fd = f(d)
    i = 0
    print("Iter\t a\t\t b\t\t c\t\t d\t\t f(c)\t\t f(d)\t\t (b - a)")
    while (b - a) > tol and i < max_iter:
        print(f"{i}\t {a:.6f}\t {b:.6f}\t {c:.6f}\t {d:.6f}\t {fc:.6f}\t {fd:.6f}\t {(b - a):.6f}") 
        if fc < fd:
            b = d
            d = c
            fd = fc
            c = a + p * (b - a)
            fc = f(c)
        else:
            a = c
            c = d
            fc = fd
            d = a + (1 - p) * (b - a)
            fd = f(d)
        i += 1

    return (a + b) / 2
```

```python
def f1(x):
    return x**4 - 14*x**3 + 60*x**2 - 70*x

minimum = golden_section(f1, 0, 2)
print("Minimum of f1 is approximately at x =", minimum)
```
