## NAIVE GAUSSIAN ELIMINATION

```python
import numpy as np
```

```python
def gauss_elimination(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)
    print("Augmented matrix awal:")
    print(np.column_stack((A, b)))

    # Forward elimination
    for i in range(n-1):
        for j in range(i+1, n):
            factor = A[j, i] / A[i, i]
            A[j] -= factor * A[i]
            b[j] -= factor * b[i]
            print(f"R{j+1} = R{j+1} - ({factor:.3f})R{i+1}")
            print(np.column_stack((A, b)))

    # Back Substitution V1 Sesuai Pseudocode
    x = np.zeros(n)
    x[n-1] = b[n-1] / A[n-1][n-1] # Elemen terakhir
    for i in range(n-2, -1, -1): 
        sum_val = b[i]
        for j in range(i+1, n):
            sum_val = sum_val - A[i][j] * x[j] 
        x[i] = sum_val / A[i][i]
    
    # Back substitution V2 (lebih sederhana)
    # x = np.zeros(n)
    # for i in range(n-1, -1, -1):
    #     x[i] = (b[i] - np.dot(A[i][i+1:], x[i+1:])) / A[i][i]

    
    return x
```

### Contoh 1
Diberikan sistem persamaan
$$x_1+2x_2+3x_3=8$$
$$2x_1+3x_2+2x_3=10$$
$$3x_1+x_2+2x_3=7$$
Dapat dituliskan sebagai:
$$A= \begin{bmatrix}
 1 & 2 & 3  \\
 2 & 3 & 2 \\
 3 & 1 & 2
\end{bmatrix},
b= \begin{bmatrix}
 8 \\
 10 \\
 7 \\
\end{bmatrix}
$$

```python
A1 = np.array([[1, 2, 3],
              [2, 3, 2],
              [3, 1, 2]
              ])
b1 = np.array([8, 10, 7])

x = gauss_elimination(A1, b1)
print("\nSolusi Akhir:")
print(f"x1 = {x[0]:.4f}")
print(f"x2 = {x[1]:.4f}")
print(f"x3 = {x[2]:.4f}")
```
