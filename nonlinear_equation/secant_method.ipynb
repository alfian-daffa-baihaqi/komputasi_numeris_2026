### DEFINISIKAN FUNGSI SECANT METHOD

```python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    i = 0
    error = float('inf')
    print(f"Initial guess: x0 = {x0}, x1 = {x1}")
    print("Iter\t x0\t\t x1\t\t x2\t\t f(x2)\t\t Error")
    while error > tol and i < max_iter:
        fx0 = f(x0)
        fx1 = f(x1)
        if fx1 - fx0 == 0:
            print("Error: Division by zero")
            return None
        x2 = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
        error = abs(x2 - x1)
        x0, x1 = x1, x2
        i += 1
        print(f"{i}\t {x0:.6f}\t {x1:.6f}\t {x2:.6f}\t {f(x2):.6f}\t {error:.6f}")
    return x1
```

### Contoh 1: 
* Diberikan sistem persamaan: 
$$f(x)=x^5+x^3+3$$
* Dengan initial value: 
$$x_0=-1, x_1=-1.1$$

```python
def f1(x):
    return x**5 + x**3 + 3

root = secant_method(f1, -1, -1.1)
print(f"Root: {root}")
```

### Contoh 2
* Diberikan sistem persamaan: 
$$f(x)=x-cos(x)$$
* Dengan initial value: 
$$x_0=0.6, x_1=0.8$$

```python
def f2(x):
    return x - np.cos(x) 

root = secant_method(f2,0.6, 0.8, tol=1e-13)
print(f"Root: {root}")
```
