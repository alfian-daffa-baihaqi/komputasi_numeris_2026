### Definisikan Fungsi False Position Method

```python
def false_position_method(f, x1, x2, tol=1e-6, max_iter=100):
    i = 0
    error = float('inf')
    print(f"Initial interval: x1 = {x1}, x2 = {x2}")
    print("Iter\t x1\t\t x2\t\t x_new\t\t f(x_new)\t Error")
    while error > tol and i < max_iter:
        f1 = f(x1)
        f2 = f(x2)
        if f1 * f2 >= 0:
            print("Nilai fungsi pada kedua sama! Pilih interval yang berbeda.")
            return None
        x_new = (x1 * f2 - x2 * f1) / (f2 - f1)
        error = abs(f(x_new))
        print(f"{i}\t {x1:.6f}\t {x2:.6f}\t {x_new:.6f}\t {f(x_new):.6e}\t {error:.6e}")
        if f1 * f(x_new) < 0:
            x2 = x_new
        else:
            x1 = x_new
        i += 1
    return x_new

```

#### Contoh 1: 
* Diberikan sistem persamaan: 
$$x^3 - 0.165x^2 + 3.993\times10^{-4}$$
* Dengan initial value: 
$$x_0=1, x_2=0.11$$

```python
def f1(x):
    return x**3 - 0.165*x**2 + 3.993*10**(-4)
root = false_position_method(f1, 0, 0.11, tol=1e-13)
print("Akar persamaan nonlinear:", root)
```
