# Tugas: Jacobi Iteration & Gauss-Seidel
Suatu persamaan linear: 
$$x+8y-2z=-9$$
$$-2x+4y-9z=-50$$
$$10x-3y+6z=24,5$$
Dapat dituliskan sebagai:
$$A= \begin{bmatrix}
 1 & 8 & -2  \\
 -2 & 4 & -9 \\
 10 & -3 & 6
\end{bmatrix},
b= \begin{bmatrix}
 -9 \\
 -50 \\
 24,5 \\
\end{bmatrix}
$$
Hitung nilai $x, y,$ dan $z$ dengan menggunakan:
* a. Metode eliminasi gauss
* b. Metode Gauss Jordan
* c. Metode Iterasi Jacobi
* d. Metode Iterasi Gauss-Seidel

```python
# LIBRARY
import numpy as np
# DEFINISIKAN MATRIKS
A = np.array([
    [1, 8, -2],
    [-2, 4, -9],
    [10, -3, 6]], dtype=float)
b = np.array([-9, -50, 24.5], dtype=float)
```

## a. Metode Eliminasi Gauss

```python
def gauss_elimination(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)
    print("Augmented matrix awal:")
    print(np.column_stack((A, b)))

    # Forward elimination
    for i in range(n-1):
        for j in range(i+1, n):
            factor = A[j, i] / A[i, i]
            A[j] -= factor * A[i]
            b[j] -= factor * b[i]
            print(f"R{j+1} = R{j+1} - ({factor:.3f})R{i+1}")
        print(np.column_stack((A, b)))

    # Back Substitution V1 Sesuai Pseudocode
    x = np.zeros(n)
    x[n-1] = b[n-1] / A[n-1][n-1] # Elemen terakhir
    for i in range(n-2, -1, -1): 
        sum_val = b[i]
        for j in range(i+1, n):
            sum_val = sum_val - A[i][j] * x[j] 
        x[i] = sum_val / A[i][i]
    
    return x
```

```python
x = gauss_elimination(A, b)
print("\nSolusi Akhir:")
print(f"x1 = {x[0]:.4f}")
print(f"x2 = {x[1]:.4f}")
print(f"x3 = {x[2]:.4f}")
```

## b. Metode Gauss-Jordan

```python
def gauss_jordan_elimination(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)

    print("Augmented matrix awal:")
    print(np.column_stack((A, b)))

    for i in range(n):
        # Normalisasi pivot
        pivot = A[i, i]
        A[i] /= pivot
        b[i] /= pivot
        print(f"\nR{i+1} = R{i+1} / {pivot}")

        # Eliminasi baris lain
        for j in range(n):
            if j != i:
                factor = A[j, i]
                A[j] -= factor * A[i]
                b[j] -= factor * b[i]
                print(f"R{j+1} = R{j+1} - ({factor})R{i+1}")

        print(np.column_stack((A, b)))

    print("\nHasil akhir:")
    print(np.column_stack((A, b)))

    return b
```

```python
x = gauss_jordan_elimination(A, b)
print("\nSolusi Akhir:")
print(f"x1 = {x[0]:.4f}")
print(f"x2 = {x[1]:.4f}")
print(f"x3 = {x[2]:.4f}")
```

## c. Metode Iterasi Jacobi

```python
def diagonally_dominant(A,b):
    A = A.copy()
    b = b.copy()
    n = len(b) 
    for i in range(n):
        max_row = i + np.argmax(np.abs(A[i:, i]))
        if max_row != i:
            A[[i, max_row]] = A[[max_row, i]]
            b[[i, max_row]] = b[[max_row, i]]
    print("Setelah diagonally dominant:")
    print(np.column_stack((A, b)))
    return A,b
```

```python
def jacobi_iteration(A, b, x, tol=1e-2, max_iter=100):
    print("Matirks Awal:")
    print(np.column_stack((A,b)))
    n = len(b)
    #swap row agar diagonally dominant
    A, b = diagonally_dominant(A,b)
    print(f"{'Iter':>4} | {'x1':>8} {'x2':>8} {'x3':>8} | {'%ex1':>8} {'%ex2':>8} {'%ex3':>8}")
    for k in range(max_iter):
        x_new = np.zeros(n)
        for i in range(n):
            s = np.sum(A[i,:] * x) - A[i,i] * x[i] # Jumlahan aij*xj kecuali diagonalnya. 
            x_new[i] = (b[i] - s) / A[i,i] #simpan di x_new
        
        # Hitung Error
        abs_error = np.abs(x_new - x)
        rel_er = (abs_error / np.abs(x_new)) * 100
        print(f"{k+1:4d} | {x_new[0]:8.4f} {x_new[1]:8.4f} {x_new[2]:8.4f} | "f"{rel_er[0]:8.4f} {rel_er[1]:8.4f} {rel_er[2]:8.4f}")
        if np.max(abs_error) < tol:
            print("\nConverged.")
            return x_new
        #update
        x = x_new
    return x
```

```python
#Nilai tebakan awal
x0 = np.zeros(3)
solution = jacobi_iteration(A, b, x0)

print("\nSolusi:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```

## d. Metode Gauss-Seidel

```python
def diagonally_dominant(A,b):
    A = A.copy()
    b = b.copy()
    n = len(b)
    for i in range(n):
        max_row = i + np.argmax(np.abs(A[i:, i]))
        if max_row != i:
            A[[i, max_row]] = A[[max_row, i]]
            b[[i, max_row]] = b[[max_row, i]]
    print("Diagonally dominant:")
    print(np.column_stack((A, b)))
    return A,b
```

```python
# Gauss Seidel
def gauss_seidel_iteration(A, b, x, tol=1e-2, max_iter=100):
    n = len(b)
    A, b = diagonally_dominant(A,b)
    print(f"{'Iter':>4} | {'x1':>8} {'x2':>8} {'x3':>8} | {'%ex1':>8} {'%ex2':>8} {'%ex3':>8}")

    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            s = np.sum(A[i,:] * x) - A[i,i] * x[i]
            x[i] = (b[i] - s) / A[i,i] #simpan di x, langsung di pakai iter selanjtunya.

        # hitung error
        abs_error = np.abs(x - x_old)
        rel_er = (abs_error / np.abs(x)) * 100
        print(f"{k+1:4d} | {x[0]:8.4f} {x[1]:8.4f} {x[2]:8.4f} | "f"{rel_er[0]:8.4f} {rel_er[1]:8.4f} {rel_er[2]:8.4f}")
        if np.max(abs_error) < tol:
            print("\nConverged.")
            return x

    return x
```

```python
#Nilai tebakan awal
x0 = np.zeros(3)
solution = gauss_seidel_iteration(A, b, x0)

print("\nSolusi:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```
