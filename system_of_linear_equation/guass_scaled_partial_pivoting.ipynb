# Gauss Elimintaion with Scaled Partial Pivoting

```python
import numpy as np

def gaussian_elimination_spp(A, b):
    n = len(b)
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    L = np.arange(n) #L=[0, 1, 2, 3, ...]
    S = np.max(np.abs(A), axis=1)
    print(np.column_stack((A, b)))

    for k in range(n - 1):
        # 1. Pemilihan Poros
        ratios = np.abs(A[L[k:], k]) / S[L[k:]]
        best_row_index_relative = np.argmax(ratios)
        best_row_index_absolute = best_row_index_relative + k
        L[k], L[best_row_index_absolute] = L[best_row_index_absolute], L[k]
        print(
            f"Kolom {k+1}\n" f"Rasio: {ratios}\n"
            f"Pivot: {L[best_row_index_absolute]+1} (rasio: {ratios[best_row_index_relative]})\n"
            f"Swap baris {L[k]+1} dengan baris {L[best_row_index_absolute]+1}\n"
            f"Vektor L setelah swap: {L+1}\n"
        )
        # 2. Proses Eliminasi
        for i in range(k + 1, n):
            multiplier = A[L[i], k] / A[L[k], k]
            for j in range(k, n):
                A[L[i], j] = A[L[i], j] - (multiplier * A[L[k], j])
            b[L[i]] = b[L[i]] - (multiplier * b[L[k]])
        print(np.column_stack((A, b)))
        
    # 3. Back Sustitution
    x = np.zeros(n)
    x[n - 1] = b[L[n - 1]] / A[L[n - 1], n - 1]
    for i in range(n - 2, -1, -1):
        sum_ax = 0
        for j in range(i + 1, n):
            sum_ax += A[L[i], j] * x[j]
        x[i] = (b[L[i]] - sum_ax) / A[L[i], i]
    return x, L, S
```

### Contoh 1
Diberikan sistem persamaan dalam bentuk matriks
$$
\begin{bmatrix}
1 & -1 & 2 & 1 \\
3 & 2 & 1 & 4 \\
5 & 8 & 6 & 3 \\
4 & 2 & 5 & 3
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
x_3\\
x_4
\end{bmatrix}
=
\begin{bmatrix}
1\\
1\\
1\\
-1
\end{bmatrix}
$$

```python
A2 = [
    [1, -1, 2, 1],
    [3, 2, 1, 4],
    [5, 8, 6, 3],
    [4, 2, 5, 3]
]
b2 = [1, 1, 1, -1]

# Menjalankan fungsi
x, L, S = gaussian_elimination_spp(A2, b2)

# Menampilkan Hasil
print("=== HASIL PERHITUNGAN ===")
print(f"Vektor Skala S: {S}")
# Tambah 1 ke L agar tampilannya (1-based) sama dengan di gambar
print(f"Vektor Indeks L Akhir: {L + 1}") 
print("\nSolusi Akhir:")
print(f"x1 = {x[0]:.4f}")
print(f"x2 = {x[1]:.4f}")
print(f"x3 = {x[2]:.4f}")
print(f"x4 = {x[3]:.4f}")
```
